```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
candyData = pd.read_csv("candyhierarchy2017.csv")
for col in candyData.columns:
    print(col, type(col))
```

```python
#Shape of the unchanged dataset
candyData.shape
```

```python
candyData.head()
```

```python
#Ch.7 
#Remove duplicates
candyData.drop_duplicates(inplace = True)
#Shape of the dataset with duplicate rows dropped
candyData.shape
#Shows that there were no duplicate rows
```

```python
#Ch. 7 
#Remove NaN values
noNanRows = candyData.dropna()
noNanRows.shape
#From this, we can see that every row does have NaN values, this will
#require further cleaning or to at least will need to be accounted for
#when doing any analysis.

#Replace NaN values with the most common for that column
for col in candyData.columns[1:]:
    candyData = candyData.fillna(candyData[col].value_counts().index[0])
candyData.head()
```

```python
#Ch. 8
#Merge candyData and secondData on the internal ID column

#Split the dataframe in two halves (column-wise) and give them a common field
numOfCols = len(candyData.columns)
df2 = candyData.iloc[:, numOfCols/2:]
df1 = candyData.iloc[:, :numOfCols/2]
df2['Internal ID'] = candyData['Internal ID']

#Merge the two dataframes back together to make the dataframe whole again.
joinedDf = df1.merge(df2, on='Internal ID', how='outer')
joinedDf.shape
```

```python
#Ch. 10
#Apply

def timesTwo(x):
    return x*2

joinedDf['Internal ID'] = joinedDf['Internal ID'].apply(timesTwo)
joinedDf.head()
```

```python
#Ch. 11
#Generate a data range
secondData = candyData['Internal ID'].to_frame()
dates = np.random.choice(pd.date_range('1970-01-01', '2020-10-24'), len(secondData))
secondData['dates'] = dates
secondData.head()
```

```python
#Ch. 11
#Convert datetime to string
secondData['dates'] = secondData['dates'].apply(str)
```
